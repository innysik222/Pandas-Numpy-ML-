## 4. Создание новых признаков

### 4.1. Создайте признак, который является суммой баллов по математике, чтению и написанию.

```python
# Ваш код здесь
df['sum_score']=df['reading score']+df['math score']+df['writing score']
df
#df['sum_score'].astype('int')
```

### 4.2. Разбейте студентов на 4 группы

Для этого создайте новый признак group



```python
# Ваш код здесь
df['group']=0
df
```

#### 4.2.1. Группа 1

У которых суммарный балл не менее 250. Назовите их high priority.

```python
# Ваш код здесь
condition=(df['sum_score'] >= 250)
df.loc[df['sum_score'] >= 250, 'group']='high priority'
df
```

#### 4.2.2. Группа 2

У которых суммарный балл не менее 200 и они пришли подготовительные курсы. Назовите их medium priority.

И убедитесь, что студенты с суммой более 250 баллов всё еще остались с high priority, если что поменяйте условие.

```python
# Ваш код здесь
condition2=(df['test preparation course']=='completed') & (df['sum_score'].between(200,250))
df.loc[condition2, 'group']='medium priority'
df
```

#### 4.2.3. Группа 3

У которых суммарный балл не менее 200, но они не проходили подготовительные курсы. Назовите их low priority.

```python
# Ваш код здесь
condition3=(df['test preparation course']=='none') & (df['sum_score'].between(200,250))
df.loc[condition3, 'group']='low priority'
df
```

#### 4.2.4. Группа 4

А остальным проставьте метку no priority.

```python
# Ваш код здесь
df.loc[(df['group']==0), 'group']='no priority'
df
```

#### 4.2.5. Проверьте, что все студенты получили группу

```python
# Ваш код здесь
df['group'].value_counts()
```
