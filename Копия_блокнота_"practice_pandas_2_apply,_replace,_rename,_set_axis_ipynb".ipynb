## 2. Создание и изменение признаков

### 2.1. Создайте признак sum_exam, который является суммой баллов по математике, чтению и написанию через .apply()

```python
# Ваш код здесь 
def sum_exam(x):
   return x['math score']+x['reading score']+x['writing score']

df['sum_exam']=df.apply(sum_exam, axis=1)
df
```

### 2.2. Замените значения в признаке gender с female на F, с male на M

```python
# Ваш код здесь 
df.replace(
    {'gender':
            {'female':'F', 'male': 'M'}}, 
           inplace=True)
df
```

### 2.3. Уберите слово group в признаке race/ethnicity	

```python
# Ваш код здесь 
df['race/ethnicity'].unique()
df.replace({'race/ethnicity': 
            {'group A': 'A', 'group B': 'B', 'group C': 'C', 'group D': 'D', 'group E': 'E'}}, 
            inplace=True)
df['race/ethnicity'] = df['race/ethnicity'].apply(lambda x: x.replace('group ', ''))
df.head()
```

### 2.4. Добавьте признак test_completed

Он будет принимать значение 0, если курс не завершен и 1, если курс завершен

```python
# Ваш код здесь 
#def test_completed(x):
  #if x['test preparation course']=='completed':
    #x['test_completed']==1
  #else:
    #x['test_completed']==0
  #return x['test_completed']

#df['test_completed']=df.apply(test_completed, axis=1)
df['test_completed']=0
df.loc[df['test preparation course']=='completed','test_completed']=1
df.head()
```

### 2.5. Удалите признак test preparation course

```python
# Ваш код здесь 
df.drop(columns=['test preparation course'], inplace=True)
df.head()
```

### 2.6. Добавьте признак standard_lunch

Он будет принимать значение 0, если ланч урезанный и 1, если ланч стандартный

```python
# Ваш код здесь 
df['standard_lunch']=0
df.loc[df['lunch']=='standard', 'standard_lunch']=1
df.head()
```

### 2.6. Удалите признак lunch

```python
# Ваш код здесь 
df.drop(columns=['lunch'], inplace=True)
df.head()
```

## 3. Переименование

### 3.1. Создайде новый датафрейм, где во всех колонках в названиях пробел на нижнее подчеркивание

#### 3.1.1. Через rename

```python
# Ваш код здесь
def rename_cols(x):
    return x.replace(' ', '_')
    

df.rename(columns=rename_cols)
```

#### 3.1.2. Через set_axis

```python
# Ваш код здесь
df.set_axis([col.replace(' ', '_') for col in df.columns], axis=1)
```

#### 3.1.3. Через атрибут columns

```python
# Ваш код здесь
df.columns=[col.replace(' ','_') for col in df.columns]
df.head()
```
